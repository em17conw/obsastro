```python
import numpy as np
from matplotlib import pyplot as plt
import glob
%matplotlib inline
```

```python
# read in Positions-mags file
pmag=np.loadtxt('Positions-mags')
xstar=pmag[:,0]
ystar=pmag[:,1]
g=pmag[:,2]
r=pmag[:,3]
i=pmag[:,4]
```

```python
# observation times
obstime=np.array([1,2,3,4,5],'f')
```

```python
# get output from qphot
qfiles=glob.glob('rfvest*.dat')
instmag=np.zeros((len(qfiles),len(g)),'f')
instmagerr=np.zeros((len(qfiles),len(g)),'f')
for i in range(len(qfiles)):
    fdat=np.loadtxt(qfiles[i])
    instmag[i]=fdat[:,0]
    instmagerr[i]=fdat[:,1]
    
```

```python
instmag.shape
```

```python
# plot magnitude versus time
plt.figure()
magoffset=np.array([0,0,0,0,0],'f')
for i in range(instmag.shape[1]):
    #plt.plot(obstime,instmag[:,i]-np.mean(instmag[:,i])+magoffset)
    plt.errorbar(obstime,instmag[:,i]-np.mean(instmag[:,i]+magoffset),yerr=instmagerr[:,i])
```
