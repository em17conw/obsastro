# Spectroscopic Analysis of the Sun

* identify emission lines from calibration lamps
    * make arrays that contain pixel values and wavelengths
    * fit linear function to wavelengths vs pixels
* fit a polynomial to spectra.
    * subtract polynomial
* convert spectrum from pixels to wavelength
* shift sky spectrum relative to sun spectrum so that absorption features line up
* scale sky and subtract from sun spectrum


```python
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def get1dspec(data,ymin,ymax):
    #print ymin,ymax
    newdat=data[ymin:ymax,:]

    # create 1-D array of flux vs pixel column
    # weight the row by the flux
    return np.sum(newdat,axis=0)
```

```python
# identify wavelengths of calibration spectra


wavelengths=np.array([6563., #Halpha
                      4868., #Hbeta
                      7780., #Helium
                      7280.,
                      6675., #Helium
                      5874., #Helium
                      5014., #Helium
                      4470., #Helium
                      ],'f')

pixelvalues=np.array([199., #Halpha
                      594., #Hbeta
                      32., #Helium
                      82., #Helium
                      172., #Helium
                      358., #Helium
                      558.,  #Helium
                      682.],'f')

# create mapping between wavelength and emission lines
plt.figure()
plt.plot(pixelvalues,wavelengths,'bo')
plt.ylabel('wavelength')
plt.xlabel('pixel')
c1=np.polyfit(pixelvalues,wavelengths,3)
xfit=np.linspace(0,700,100)
plt.plot(xfit,np.polyval(c1,xfit),'r--')
print c1
```
