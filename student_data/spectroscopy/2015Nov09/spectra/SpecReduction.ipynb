```python
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def get1dspec(data,ymin,ymax):
    #print ymin,ymax
    newdat=data[ymin:ymax,:]

    # create 1-D array of flux vs pixel column
    # weight the row by the flux
    return np.sum(newdat,axis=0)
```

```python
# identify wavelengths of calibration spectra


wavelengths=np.array([6563., #Halpha
                      4868., #Hbeta
                      5852., #Neon
                      6598., #Neon
                      4358., #Mercury
                      5461., #Mercury
                      6506. #Neon
                      ],'f')

pixelvalues=np.array([83.5, #Halpha
                      477., #Hbeta
                      234., #Neon
                      82.75, #Neon
                      580., #Mercury
                      325., #Mercury
                      107. #Neon
                      ],'f')

# create mapping between wavelength and emission lines
plt.figure()
plt.plot(wavelengths,pixelvalues,'bo')
c1=np.polyfit(wavelengths,pixelvalues,1)
xfit=np.linspace(4000,7000,100)
plt.plot(xfit,np.polyval(c1,xfit),'r--')
```
