```python
import numpy as np
from matplotlib import pyplot as plt
import glob
%matplotlib inline
```

```python
# read in Positions-mags file
pmag=np.loadtxt('ZPositions-mags')
xstar=pmag[:,0]
ystar=pmag[:,1]
g=pmag[:,2]
r=pmag[:,3]
i=pmag[:,4]

Bknown = g + .39*(g-r) + .21
Vknown = g - 0.59*(g-r) -.01
VRknown = 1.09*(r-i) +.22
Rknown = Vknown - VRknown

# get output from qphot
Binst=np.loadtxt('Bband.dat')
Vinst=np.loadtxt('Vband.dat')
Rinst=np.loadtxt('Rband.dat')

```

```python
# plot Bknown vs Binst
plt.figure()
BZP=-2.7
plt.plot(Binst[:,0][0:-1],Bknown[0:-1],'bo')
plt.errorbar(Binst[:,0][0:-1],Bknown[0:-1],xerr=Binst[:,1][0:-1],fmt=None)
plt.xlabel('B instrumental',fontsize=14)
plt.ylabel('B from SDSS',fontsize=14)
xl=np.arange(15,18.5,.1)
plt.plot(xl,xl+BZP,'r')
plt.axis([15,18.5,11,17])
```

```python
# plot Vknown vs Vinst
plt.figure()
VZP=-3.2
plt.plot(Vinst[:,0][0:-1],Vknown[0:-1],'bo')
plt.errorbar(Vinst[:,0][0:-1],Vknown[0:-1],xerr=Vinst[:,1][0:-1],fmt=None)
plt.xlabel('V instrumental',fontsize=14)
plt.ylabel('V from SDSS',fontsize=14)
xl=np.arange(14,18.5,.1)
plt.plot(xl,xl+VZP,'r')
plt.axis([14,18.,10,16])
```

```python
# plot Rknown vs Rinst
plt.figure()
RZP=-3.4
plt.plot(Rinst[:,0][0:-1],Rknown[0:-1],'bo')
plt.errorbar(Rinst[:,0][0:-1],Rknown[0:-1],xerr=Rinst[:,1][0:-1],fmt=None)
plt.xlabel('R instrumental',fontsize=14)
plt.ylabel('R from SDSS',fontsize=14)
xl=np.arange(14,18.5,.1)
plt.plot(xl,xl+RZP,'r')
plt.axis([14,18.,10,16])
```

```python
# plot color-color diagram of asteroids and stars
B = Binst[:,0]+BZP

R = Rinst[:,0]+RZP
V = Vinst[:,0]+VZP
Berr = Binst[:,1]
Verr = Vinst[:,1]
Rerr = Rinst[:,1]
VR = V - R
BV = B - V
VRerr=np.sqrt(Verr**2 + Rerr**2)
BVerr=np.sqrt(Verr**2 + Berr**2)
VR=VR[Rknown < 16]
BV=BV[Rknown < 16]
VRerr=VRerr[Rknown < 16]
BVerr=BVerr[Rknown < 16]
plt.figure()
plt.plot(VR[0:-1],BV[0:-1],'b*',markersize=12,label='Stars')
plt.xlabel('V-R',fontsize=14)
plt.ylabel('B-V',fontsize=14)
plt.plot(VR[-1:],BV[-1:],'rs',markersize=16,label='Vesta')
plt.legend(loc='upper right',numpoints=1)
plt.axhline(y=.76,color='k',ls='--') # B-V > 0.76 for S type
plt.axvline(x=.09,color='k',ls='--') # V-R > 0.09 for S type
```
